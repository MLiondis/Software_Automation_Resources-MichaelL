__7. Refering to the below model, describe underfitting in Linear Regression.__

Underfitting is when a model is too simple to learn the underlying patterns in the training data, resulting in poor performance on both training and new, unseen data. The model below shows a linear regression function that lies well above the training data points, meaning it is underfitting the data and will produce results that are very inaccurate. 

````python
x_train = np.array([1, 1.5, 1.8, 2, 2.2])
y_train = np.array([[200], [250], [260], [280], [500]])
tmp_f_wb = compute_model_output(x_train, 200, 10,)
plt.plot(x_train, tmp_f_wb, c='b',label='Our Prediction')
plt.scatter(x_train, y_train, marker='x', c='r',label='Actual Values')
plt.show()
````
